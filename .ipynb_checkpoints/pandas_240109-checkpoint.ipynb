## 데이어 다뤄보기
- 행과 열을 다뤄보기

```python
# 판매량 칼럼 추가하기
df['Sold'] = 0
df.head(1)
```

```python
# Lemon과 Orange를 합쳐서 구하기
df['Sold'] = df['Lemon'] + df['Orange']
df
```

```python
# 매출액 구하기
df['매출액'] = df['Sold'] * df['Price']
df = df.drop_duplicates(keep = 'first', ignore_index=True)
df.head(1)
```

## 분석 프로젝트

- 판매량, 단가
- 판매량 (O) / 단가 (X) => 단가 리스트를 구하는 것이 프로젝트

## GroupBy 연산
- SQL : GROUP By와 Logic은 동일함
- Groupby 기준칼럼 : 패턴이 있는 문자열 => 그룹화
- Groupby 대상칼럼 : 수치 데이터
- 필요한 함수 : 집계함수 (합, 평균, 표준편차, 최대, 최소 등)

```python
# df.groupby('기준칼럼')['대상칼럼'].agg(['집계함수'])
df.groupby('Location')['매출액'].agg(['sum'])
```

```python
df.groupby('Location')['매출액'].agg(['sum','mean'])
```

```python
# reset_index()를 사용하면 'Location'이 인덱스로 들어간게 열로 다시 오게 된다.
df.groupby('Location')['매출액'].agg(['sum','mean']).reset_index()
```
