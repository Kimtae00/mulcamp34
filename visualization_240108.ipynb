## 라이브러리 불러오기

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# tips 데이터 불러오기
tips = sns.load_dataset('tips')

# 도화지 그리기
fig, ax = plt.subplots()

# matplotlib 시각화
ax.plot([1, 2, 3])

# 산점도 그래프 그리기
# data : tips 라는 pandas DataFrame을 받는다
# x축 : '컬럼명' = total_bill
# y축 : '컬럼명' = tip

# ax는 seaborn과 matplotlib를 이어주는 매개변수 역할을 한다고 생각하면 된다.
# 모든 것이 제공해주는 것은 아니므로 확인을 해봐야 한다.
# ax.legend()로 범례를 만들어야 하는 것과는 달리 hue 옵션을 주기만 해도 범례가 만들어진다.
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax = ax, data = tips) # seaborn 시각화

# output 폴더에 result.png로 그래프 저장하기
plt.savefig('output/result.png')

# 그래프 그리기
# 가장 마지막에 작성하는거 중요 !
plt.show() 
```

## matplotlib와 seaborn을 동시에 사용한 subplotting 기법
- 다중차트를 하나의 도화지에 담기

```python
# 도화지 그리기
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# 각 차트 위에 타이틀 지정하기
ax[0][0].set_title('ax[0][0] title')
ax[1][1].set_title('ax[1][1] title')
ax[0][1].set_title('ax[0][1] title')
ax[1][0].set_title('ax[1][0] title')

# 0,0 위치
ax[0][0].plot([1, 2, 3])

# 0,1 위치
x = np.arange(-5, 5, 0.1)
y1 = x**2 - 2
y2 = 20*np.cos(x)**2

ax[0][1].plot(x, y2)

# 1,0 위치
ax[1][0].plot(x, y1)

# 1,1 위치
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax = ax[1][1], data = tips) # seaborn 시각화

# 가운데에 타이틀 지정
fig.suptitle('fig title', size=15)

# 간격 띄우기
plt.tight_layout()

# 그래프 출력하기
plt.show()
```
